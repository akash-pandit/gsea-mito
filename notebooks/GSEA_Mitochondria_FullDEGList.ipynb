```python
deg_files = ["deseq.results.allGenes.hscxy.hscxo.txt","deseq.results.allGenes.mkpgo.mkpto.txt","deseq.results.allGenes.mkpgy.mkpgo.txt","deseq.results.allGenes.mkpgy.mkpxo.txt","deseq.results.allGenes.mkpgo.hscxo.txt","deseq.results.allGenes.mkpgy.hscxy.txt","deseq.results.allGenes.mkpgy.mkpto.txt","deseq.results.allGenes.mkpto.hscxo.txt"]
for deg_file in deg_files: 
    print("DEG_LISTS/"+deg_file)
    deg_df = pd.read_csv("DEG_LISTS/"+deg_file, sep=r"\s+")
    # deg_df = pd.read_csv("DEG_LISTS/deseq.results.hscxy.hscxo.txt", sep=r"\s+")
    deg_df['symbol'] = deg_df['symbol'].astype(str).str.upper() # uppercase the gene name
    deg_df_clean = deg_df[deg_df['symbol'] != "SYMBOL"]
    # deg_df_clean_filtered = deg_df_clean[(deg_df_clean['baseMean_hscxy'].astype(float) > 10) & (deg_df_clean['baseMean_hscxo'].astype(float) > 10)]
    deg_df_clean['log2FcMAP'] = pd.to_numeric(
        deg_df_clean['log2FcMAP'],
        errors='coerce'
    )
    
    deg_df_clean['padj'] = pd.to_numeric(
        deg_df_clean['padj'],
        errors='coerce'
    )
    rnk = deg_df_clean.sort_values(ascending=False, by="log2FcMAP")[['symbol', 'log2FcMAP']].set_index("symbol") # Created Ranked list

    #####################
    ##### GSEA PLOT #####
    #####################
    pre_res = gp.prerank(rnk=rnk, # or rnk = rnk,
                     gene_sets='MSigDB_Hallmark_2020',
                     threads=4,
                     min_size=5,
                     max_size=1000,
                     permutation_num=1000, # reduce number to speed up testing
                     outdir=None, # don't write to disk
                     seed=6,
                     verbose=True, # see what's going on behind the scenes
                    )

    terms = pre_res.res2d.Term # GSEA output

    # Oxidative Phosphorylation
    # ox_phos_index = terms[terms == "Oxidative Phosphorylation"].index[0] #Find the Oxphos index
    # axs = pre_res.plot(terms[ox_phos_index], ofname="GSEA_OXPHOS_" + deg_file + ".fullGeneList.pdf") ### Save OxPhos Figure

    # ROS 
    # ros_index = terms[terms == "Reactive Oxygen Species Pathway"].index[0] # ROS index 
    # axs = pre_res.plot(terms[ros_index], ofname="GSEA_ROS_" + deg_file + ".fullGeneList.pdf") ### Save ROS Figure

    # TBD 
    term_index = terms[terms == "heme Metabolism"].index[0] # Term index
    axs = pre_res.plot(terms[term_index], ofname="GSEA_HemeMetabolism_" + deg_file + ".fullGeneList.pdf") ### Save Figure

    ########################
    ##### Volcano Plot #####
    ########################
    # Load your DESeq2 results
    df = deg_df_clean.set_index("symbol")
    
    # Filter DESEQ2 for genes in GO Term 
    df = df[df.index.isin(ox_phos_genes)]
    
    # Drop rows with missing values (optional)
    df = df.dropna(subset=["padj", "log2FcMAP"])
    
    # Define significance thresholds
    padj_thresh = 0.05
    lfc_thresh = 0.5
    
    # Add significance column for coloring
    df["significance"] = "Not sig."
    df.loc[(df["padj"] < padj_thresh) & (df["log2FcMAP"] > lfc_thresh), "significance"] = "Up"
    df.loc[(df["padj"] < padj_thresh) & (df["log2FcMAP"] < -lfc_thresh), "significance"] = "Down"
    
    # Create the volcano plot
    plt.figure(figsize=(8,6))
    plt.scatter(df["log2FcMAP"], -np.log10(df["padj"]),
                c=df["significance"].map({"Up":"red", "Down":"blue", "Not sig.":"grey"}),
                alpha=0.7, s=10)
    
    # Add threshold lines
    plt.axhline(-np.log10(padj_thresh), color='black', linestyle='--', lw=1)
    plt.axvline(x=lfc_thresh, color='black', linestyle='--', lw=1)
    plt.axvline(x=-lfc_thresh, color='black', linestyle='--', lw=1)
    
    # Labels
    plt.xlabel("log2(Fold Change)")
    plt.ylabel("-log10(adjusted p-value)")
    plt.title("Volcano Plot")
    # plt.set_ylim(20)
    df['gene'] = df.index
    # Optional: annotate top genes
    if "gene" in df.columns:
        top_genes = df.sort_values("padj").head(25)
        for _, row in top_genes.iterrows():
            plt.text(row["log2FcMAP"], -np.log10(row["padj"]), row["gene"],
                     fontsize=8, ha='right' if row["log2FcMAP"] < 0 else 'left')
    # plt.ylim(0, 10)
    
    plt.tight_layout()
    # plt.savefig("lung.ILC2.pb.volcano.png")
    # plt.show()
    plt.title(deg_file+" heme Metabolism Genes")
    plt.savefig("GSEA_hemeMetabolism_" + deg_file + ".fullGeneList.volcanoPlot.png")
```
